# **1. MEJOR Y PEOR CASO:**  s
Sabemos que hay $n$ listas con $n$ entradas , $n$ mujeres y $n$ hombres , sabemos que ningun hombre puede ser rechazado $n$ veces , porque si fuera rechazado $n$ veces se quedaria sin pareja , este algoritmo garantiza que siempre se puede formar una pareja estable es decir no esposible que ningun hombre se quede sin una pareja a quien proponerle matrimonio, entoces el pero de los casos seria que un hombre fuera rechazado todos los dias hasta que finalice el algoritmo.


Podemos decir que hay $n$ hombres con maximo $n-1$ rechazos cada uno, es decir que hay un maximo de n hombres y hay un total  $n(n-1)$ propuestas rechazadas; ahora suponga que en la $n(n-1)$ propuesta el hombre x es rechazado por n-1 vez, en ese caso todavia nesecitaria una propuesta adicional para proponerle matrimonio a una mujer de su preferencia y en ese caso el algoritmo de matrimonio estable termina, esto significa que debe haber exactemente un dia en el que no se hagan propuestas. 
Por esto podemos concluir que el peor caso del algoritmo de matrimonio estable es la suma del maximo de rechazos que puede tener un hombre mas el momento en que ninguna propuesta puede ser rechazado. Es decir $n(n-1)+1$


El mejor de los casos , es en el que $n$ hombres son aceptados al hacer la primera propuesta; Es decir $n$ pasos. 


**Ejemplo:** 
En este algoritmo para $n=4$ tenemos que el mejor caso es 4, y el peor caso es $4(4-1)+1=13$

Asi mismo para $n=40$ tenemos que el mejor caso es 40, y el peor caso es $40(40-1)+1=1561$


```python
def bad(n):
  return (n*(n-1)+1)

for i in range (4,41):

  print("n = ",i,"Pn = ", bad(i))
```

El algoritmo de ***Irving*** determina si existe una coincidencia estable, y de encontrarla, será con una complejidad de O(n^2). Para el mejor de los casos, también se considera que se encuentre una coincidencia estable para cada miembro en la primera iteración, por lo que consideraríamos un mínimo de pasos n

```python
def badIrving(n):
  return (n*n)

for i in range (4,41):

  print("n = ",i,"Pn = ", badIrving(i))
```

```python


import random
import statistics as std
import matplotlib.pyplot as plt

def createPreference(N):
  prefer = []
  for i in range(1, 2*N+1):
    preference = []
    while (len(preference) < N):
      if len(prefer)<(N):
        k = random.randint(1, N)
        if k in preference:
          continue
        preference.append(k)
      else:
        k = random.randint(N+1, 2*N)
        if k in preference:
          continue
        preference.append(k)
    prefer.append(preference)
  return prefer
```

```python
def algoCosa(N):
  prefer=createPreference(N)  
  # This function returns true if  
  # woman 'w' prefers man 'm1' over man 'm' 
  def wPrefersM1OverM(prefer, w, m, m1): 
        
      # Check if w prefers m over her  
      # current engagment m1 
      for i in range(N): 
            
          # If m1 comes before m in lisr of w,  
          # then w prefers her current engagement, 
          # don't do anything 
          if (prefer[w][i] == m1): 
              return True
    
          # If m cmes before m1 in w's list,  
          # then free her current engagement  
          # and engage her with m 
          if (prefer[w][i] == m): 
              return False
    
  # Prints stable matching for N boys and N girls.  
  # Boys are numbered as 0 to N-1.  
  # Girls are numbereed as N to 2N-1. 
  def stableMarriage(prefer): 
        
      # Stores partner of women. This is our output  
      # array that stores paing information.  
      # The value of wPartner[i] indicates the partner  
      # assigned to woman N+i. Note that the woman numbers  
      # between N and 2*N-1. The value -1 indicates  
      # that (N+i)'th woman is free 
      wPartner = [-1 for i in range(N)] 
    
      # An array to store availability of men.  
      # If mFree[i] is false, then man 'i' is free, 
      # otherwise engaged. 
      mFree = [False for i in range(N)] 
    
      freeCount = N 
    
      # While there are free men 
      while (freeCount > 0): 
            
          # Pick the first free man (we could pick any) 
          m = 0
          while (m < N): 
              if (mFree[m] == False): 
                  break
              m += 1
    
          # One by one go to all women according to  
          # m's preferences. Here m is the picked free man 
          i = 0
          while i < N and mFree[m] == False: 
              w = prefer[m][i] 
    
              # The woman of preference is free,  
              # w and m become partners (Note that  
              # the partnership maybe changed later).  
              # So we can say they are engaged not married 
              if (wPartner[w - N] == -1): 
                  wPartner[w - N] = m 
                  mFree[m] = True
                  freeCount -= 1
    
              else:  
                    
                  # If w is not free 
                  # Find current engagement of w 
                  m1 = wPartner[w - N] 
    
                  # If w prefers m over her current engagement m1, 
                  # then break the engagement between w and m1 and 
                  # engage m with w. 
                  if (wPrefersM1OverM(prefer, w, m, m1) == False): 
                      wPartner[w - N] = m 
                      mFree[m] = True
                      mFree[m1] = False
              i += 1
    
              # End of Else 
          # End of the for loop that goes  
          # to all women in m's list 
      # End of main while loop 
    
      # Prthe solution 
      print("Woman ", " Man") 
      for i in range(N): 
          print(i + N, "\t", wPartner[i]) 
    
  # Driver Code 
  
  print(prefer)  
  return stableMarriage(prefer)

print("Parejas estables para n=4 \n")
algoCosa(4)
print("Parejas estables para n=40 \n")
algoCosa(40)
```
