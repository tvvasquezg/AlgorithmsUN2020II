# RECURSIVE FIBONACCI 

```
def fibR(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)
```

# ITERATIVE FIBONACCI 

```
def fib(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

#ITERATIVE FIBONACCI with memory

```
def fibM(n):
    memo = {}
    def fib(n):
        if n==0 :
            return 0
        if n==1 :
            return 1
        if (n-2) not in memo:
            memo[n-2] = fib(n-2)
        if (n-1) not in memo:
            memo[n-1] = fib(n-1)
        return memo[n-1] + memo[n-2]
    return fib(n)
```

```
import time

time1=[]
for i in range(0,51):
  n=i
  t0 = time.clock()
  k = fibR(n)
  tf =time.clock()
  time1.append(tf - t0)
print(type(time1))
```
